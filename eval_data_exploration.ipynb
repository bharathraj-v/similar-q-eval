```python
sim_df = pd.read_csv("evals/similar_questions_evaluation_results.csv")

sim_df.head()
```

```python
sim_df = sim_df[sim_df['error'].isna()]
sim_df.head()
```

```python
def aggregate_similar(df):
    return pd.Series({
        "similar_questions_ns": df["similar_question_text"].tolist(),
        "avg_original_similarity_score_ns": df["original_similarity_score"].mean(),
        "avg_conceptual_similarity_score_ns": df["conceptual_similarity_score"].mean(),
        "avg_structural_similarity_score_ns": df["structural_similarity_score"].mean(),
        "avg_difficulty_alignment_score_ns": df["difficulty_alignment_score"].mean(),
        "avg_solution_approach_transferability_score_ns": df["solution_approach_transferability_score"].mean(),
        "avg_total_score_ns": df["total_score"].mean(),
        "avg_processing_time_ns": df["processing_time"].mean(),
        "evaluation_notes_ns": " | ".join(df["evaluation_notes"].dropna().unique().tolist())
    })

sim_metrics_ns_df = sim_df.groupby(
    ["question_id", "original_question", "subject"], as_index=False
).apply(aggregate_similar).reset_index()

sim_metrics_ns_df.head()
```

```python
df.head()
```

```python
merged_df = comparision_eval.merge(
    sim_metrics_df,
    on=["question_id", "subject"],  # merge keys
    how="inner"
)

merged_df.head()
```

```python
merged_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# ---- Scatter with regression lines ----
fig, axes = plt.subplots(1, 2, figsize=(12,5), sharey=True)

sns.regplot(
    data=merged_df,
    x="avg_total_score", y="sim_score",
    scatter_kws={"alpha":0.6}, line_kws={"color":"red"}, ax=axes[0]
)
axes[0].set_title("Avg Total Similarity vs Sim Score")
axes[0].set_xlabel("Avg Total Similarity Score")
axes[0].set_ylabel("Score")

sns.regplot(
    data=merged_df,
    x="avg_total_score", y="standard_score",
    scatter_kws={"alpha":0.6}, line_kws={"color":"blue"}, ax=axes[1]
)
axes[1].set_title("Avg Total Similarity vs Standard Score")
axes[1].set_xlabel("Avg Total Similarity Score")
axes[1].set_ylabel("")

plt.tight_layout()
plt.show()


# ---- Boxplots across bins ----
merged_df["score_bin"] = pd.cut(
    merged_df["avg_total_score"],
    bins=[0,50,70,85,95,100],
    labels=["<50","50-70","70-85","85-95","95-100"]
)

fig, axes = plt.subplots(1, 2, figsize=(12,5), sharey=True)

sns.boxplot(
    data=merged_df, x="score_bin", y="sim_score", ax=axes[0]
)
axes[0].set_title("Sim Score Distribution by Similarity Bucket")
axes[0].set_xlabel("Avg Total Similarity Score Bucket")
axes[0].set_ylabel("Score")

sns.boxplot(
    data=merged_df, x="score_bin", y="standard_score", ax=axes[1]
)
axes[1].set_title("Standard Score Distribution by Similarity Bucket")
axes[1].set_xlabel("Avg Total Similarity Score Bucket")
axes[1].set_ylabel("")

plt.tight_layout()
plt.show()
```
